```python
from bs4 import BeautifulSoup
import requests
import csv 
import nltk
#from newspaper import Article
import newspaper
```

```python
#url= "https://www.cnbc.com/politics/"
url="hhttps://www.cnbc.com/politics/"
html=requests.get(url).text
soup=BeautifulSoup(html)
```

```python
news_box=soup.find_all('a',{'class': 'Card-mediaContainer'})
i=0
for news in news_box:
    print(news['href'])
    print()
    i=i+1
print(i)
```

```python
for news in news_box:
    url_new=news['href']
    article=newspaper.Article(url_new)
    article.download()
    article.parse()
    nltk.download('punkt')
    article.nlp()
    rowsData=[article.authors,article.publish_date.date().strftime("%d-%m-%Y"),url_new,article.text,article.title]
    with open(filename, 'a+') as csvfile:
        csvwriter=csv.writer(csvfile)
        csvwriter.writerow(rowsData)
```
