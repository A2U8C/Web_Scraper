```python
from bs4 import BeautifulSoup
import requests
import csv 
import nltk
import newspaper
```

```python
url= "https://www.cnbc.com/politics/"
```

```python
web_paper = newspaper.build(url)

for article in web_paper.articles:
    print(article.url)
    print()
```

```python
from re import search 
filename = "news_dataset_nonVideo.csv"
video_text="/video"
for article_url in web_paper.articles:
    if(article_url.url.find(video_text) is -1):
        try:
            article=newspaper.Article(article_url.url)
            article.download()
            article.parse()
            nltk.download('punkt')
            article.nlp()
            rowsData=[article.authors,article.publish_date.date().strftime("%d-%m-%Y"),article_url.url,article.text,article.title]
            with open(filename, 'a+') as csvfile:
                csvwriter=csv.writer(csvfile)
                csvwriter.writerow(rowsData)
        except:
            print("An exception occurred")
```

```python
from re import search 

url= "https://www.cnbc.com/politics/"
web_paper = newspaper.build(url)
filename = "news_dataset_CNBC_News.csv"
```

```python
fields = ['Authors', 'Time', 'URL', 'Contents', 'Title'] 
with open(filename, 'w') as csvfile:
    csvwriter=csv.writer(csvfile)
    csvwriter.writerow(fields)
```

```python
video_text="/video"
for article_url in web_paper.articles:
    if(article_url.url.find(video_text) is -1):
        try:
            article=newspaper.Article(article_url.url)
            article.download()
            article.parse()
            nltk.download('punkt')
            article.nlp()
            total_no=len(article.authors)
            name_sec=article.authors[0]
            #print(name_sec)
            for x in range(1, total_no):
                name_sec=name_sec+', '+article.authors[x]
            rowsData=[name_sec,article.publish_date.date().strftime("%d-%m-%Y"),article_url.url,article.text,article.title]
            with open(filename, 'a+') as csvfile:
                csvwriter=csv.writer(csvfile)
                csvwriter.writerow(rowsData)
        except:
            print("An exception occurred")
```
